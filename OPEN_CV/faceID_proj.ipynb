```python
import face_recognition
import numpy as np
import cv2

```

```python
vectordb={}
img_list=[("Virat Kohli","virat.jpg"),
          ("Ravindra Jadeja","jadeja.jpg"),
          ("Jasprit Bhumra","Jasprit.jpg")]
img_dict=dict(img_list)
for i,j in img_list:
    img_arr=face_recognition.load_image_file(j)
    img_enc=face_recognition.face_encodings(img_arr)
    vectordb[i]=img_enc[0]
    
```

```python
input=face_recognition.load_image_file("bumra.jpg")
input_enc=face_recognition.face_encodings(input)
if input_enc:
    input_enc=input_enc[0]
    result=face_recognition.compare_faces(list(vectordb.values()),input_enc,tolerance=0.5)
    if True in result:
        r=result.index(True)
        name=list(vectordb.keys())[r]
        print(name)
        img=cv2.imread(img_dict[name])
        cv2.imshow("frame",img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print("unknown face")
else:
    print("no input")
```
