```python
from tensorflow.keras.layers import Input, Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential
```

```python
model=Sequential([Input((224,224,3)),
                  Conv2D(16,(3,3),activation="relu"),
                  MaxPooling2D((2,2)),
                  Conv2D(32,(3,3),activation="relu"),
                  MaxPooling2D((2,2)),
                  Flatten(),
                  Dense(24,activation="relu"),
                  Dense(48,activation="relu"),
                  Dense(5,activation="softmax"),])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
image_generator=ImageDataGenerator(rescale=1./255, validation_split=0.2)
```

```python
train_data=image_generator.flow_from_directory("/content/data",target_size=(224,224),batch_size=32,class_mode="sparse", subset="training",seed=42)
```

```python
val_data=image_generator.flow_from_directory("/content/data",target_size=(224,224),batch_size=32,class_mode="sparse", subset="validation",seed=42)
```

```python
train_data.class_indices
```

```python
model.fit(train_data,validation_data=val_data,epochs=10)
```

```python
train_gen=ImageDataGenerator(rescale=1./255)
val_gen=ImageDataGenerator(rescale=1./255)
test_gen=ImageDataGenerator(rescale=1./255)
train_data=train_gen.flow_from_directory("/content/dataset/train",target_size=(224,224),batch_size=32,class_mode="sparse",shuffle=True)
valid_data=val_gen.flow_from_directory("/content/dataset/val",target_size=(224,224),batch_size=32,class_mode="sparse",shuffle=False)
test_data=test_gen.flow_from_directory("/content/dataset/test",target_size=(224,224),batch_size=32,class_mode="sparse",shuffle=False)
```

```python
model.fit(
    train_data,
    validation_data=valid_data,
    epochs=10
)
```

```python

class_names = [None] * len(train_data.class_indices)
for cls, idx in train_data.class_indices.items():
    class_names[idx] = cls

print(class_names)
```

```python
prediction = model.predict(img_array)

for i, prob in enumerate(prediction[0]):
    print(class_names[i], ":", prob)
```

```python
prediction=model.predict(img_array)
predicted_index=np.argmax(prediction)
predicted_class=class_names[predicted_index]
confidence=np.max(prediction)
print(f"predicted hair type:",predicted_class)
print(f"confidence:",confidence)
```
