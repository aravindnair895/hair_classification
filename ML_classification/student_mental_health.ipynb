lebal encode

```python
lis=["Choose your gender","Marital status","Do you have Depression?","Do you have Anxiety?","Do you have Panic attack?","Did you seek any specialist for a treatment?"]
dicti={}
from sklearn.preprocessing import LabelEncoder
for i in lis:
    dicti[i]=LabelEncoder()
    df[i]=dicti[i].fit_transform(df[i])
```

```python
for col,le in dicti.items():
    print(f"{col} mapping",dict(zip(le.classes_,le.transform(le.classes_))))
```

ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
od=OrdinalEncoder(categories=[["Year 1","Year 2","Year 3","Year 4"],["0 - 1.99","2.00 - 2.49","2.50 - 2.99","3.00 - 3.49","3.50 - 4.00"]])
df[["Your current year of Study","What is your CGPA?"]]=od.fit_transform(df[["Your current year of Study","What is your CGPA?"]])
df.head()
```

```python
df["What is your course?"].unique()
```

onehot encoding

```python
dff=pd.get_dummies(df,columns=["What is your course?"],dtype="int",drop_first=True)
dff
```

```python
df["Did you seek any specialist for a treatment?"].value_counts()
```

balancing

```python
x=dff.drop(columns=["Timestamp"])
y=dff["Did you seek any specialist for a treatment?"]
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
pipe=Pipeline([("oversampling",RandomOverSampler(sampling_strategy={1:50})),
               ("undersampling",RandomUnderSampler(sampling_strategy={0:50}))])
x_sample,y_sample=pipe.fit_resample(x,y)
```

scaling

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
x_scale=minmax.fit_transform(x_sample)
```

model selection

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y_sample,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
model=[KNeighborsClassifier(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),BernoulliNB(),LogisticRegression()]
for i in model:
    print(f"{i}")
    print("*"*10)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("accuracy:",round(accuracy_score(y_pred,y_test)*100),"%")
    print(classification_report(y_pred,y_test))
    sb.heatmap(confusion_matrix(y_pred,y_test),annot=True,fmt="d")
    mlt.title(f"heatmap of {i}")
    mlt.show()
```

```python

final_model=RandomForestClassifier()
final_model.fit(x_train,y_train)
y_pred1=final_model.predict(x_test)
print("accuracy:",round (accuracy_score(y_pred1,y_test)*100))
print(classification_report(y_pred1,y_test))
sb.heatmap(confusion_matrix(y_pred,y_test),annot=True,fmt="d",cmap="Blues")
mlt.xlabel("actual")
mlt.ylabel("predict")
mlt.show()
```
